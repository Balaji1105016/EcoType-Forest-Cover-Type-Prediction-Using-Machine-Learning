| Feature Selection Method            | Limitation in my Dataset                     | Why Random Forest Better                                            |
| ----------------------------------- | ---------------------------------------------- | ------------------------------------------------------------------- |
| Pearson Correlation                 | Captures only linear relationships             | Random Forest captures complex non-linear geographical patterns     |
| Spearman Correlation                | Handles monotonic relations only               | Random Forest detects arbitrary feature interactions automatically  |
| Chi-Square Test                     | Mainly suitable for categorical features       | Random Forest handles mixed numerical and categorical features      |
| ANOVA F-Test                        | Assumes statistical independence and linearity | Random Forest learns real predictive relationships directly         |
| Variance Threshold                  | Removes low variance only                      | Random Forest measures actual predictive importance                 |
| Mutual Information                  | Shows dependency but lacks interpretability    | Random Forest provides clearer ranked feature importance            |
| Forward Selection                   | Computationally expensive                      | Random Forest faster on medium-sized tabular datasets               |
| Backward Elimination                | Slow for multiple feature combinations         | Random Forest efficiently evaluates features simultaneously         |
| Recursive Feature Elimination (RFE) | Requires repeated model training               | Random Forest inherently performs embedded feature selection        |
| Lasso Regression                    | Assumes linear relationships                   | Random Forest handles non-linear environmental relationships better |
| PCA                                 | Loses original feature meaning                 | Random Forest preserves feature interpretability for deployment     |
| t-SNE                               | Mainly visualization technique                 | Random Forest directly supports predictive modeling                 |
| Decision Tree Importance            | High overfitting risk                          | Random Forest reduces variance using ensemble learning              |
| XGBoost Importance                  | More tuning complexity                         | Random Forest simpler and stable for baseline feature selection     |


```python
# Instead of fixing the outliers for all columns in Dataset, we are identifying the essential features and fixing outliers for features alone
# Out of all methods we are going with Randomforest for the above reasons:

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

x = df.drop("Cover_Type",axis = 1)
y = df["Cover_Type"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5,stratify=y)# Using stratified sampling to handle class imbalance
rfc = RandomForestClassifier(random_state=5) # Model Creation
rfc.fit(x_train,y_train) # Model Training

#Accuraccy Check
y_predict = rfc.predict(x_test)
accuraccy = accuracy_score(y_test,y_predict)

print("Before Selection")
print(f"Accuraccy: {round(accuraccy*100,2)}%")

# Important Features
# Creating the DataFrame for Important Features
importance_df = pd.DataFrame({"Important Feturures":x.columns,"Importance": rfc.feature_importances_})
#print(importance_df)

importance_df = importance_df.sort_values(by="Importance",ascending=False)

# Barplot for Visualisation
plt.figure(figsize=(10,6))
sns.barplot(data=importance_df,x="Importance",y="Important Feturures")
plt.title("Feature Importance")



```

Inference:
1) Elevation is the most dominant factor influencing the model's decisions.
2) Distance to Roadways acts as the second most influential feature.
3) Distance to Fire Points contributes significantly as the third factor.
4) Soil Type and Wilderness Area show a moderate level impact.
5) Hillshade and Slope have the lowest importance in this model.

```python
# Taking the top 5 Features and Retraining the model and checking the accuracy score:
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

x = df[["Elevation","Horizontal_Distance_To_Roadways","Horizontal_Distance_To_Fire_Points","Soil_Type","Wilderness_Area"]]
y = df["Cover_Type"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5,stratify=y)# Using stratified sampling to handle class imbalance
rfc = RandomForestClassifier(random_state=5) # Model Creation
rfc.fit(x_train,y_train) # Model Training

#Accuraccy Check
y_predict = rfc.predict(x_test)
accuraccy = accuracy_score(y_test,y_predict)

print("After Selection")
print(f"Accuraccy: {round(accuraccy*100,2)}%")

# Important Features
# Creating the DataFrame for Important Features
importance_df = pd.DataFrame({"Important Feturures":x.columns,"Importance": rfc.feature_importances_})
#print(importance_df)

importance_df = importance_df.sort_values(by="Importance",ascending=False)

# Barplot for Visualisation
plt.figure(figsize=(10,6))
sns.barplot(data=importance_df,x="Importance",y="Important Feturures")
plt.title("Feature Importance")
```

Note: Taking top 5 features out 11 and checked the Accuraccy score the same 95 percent it is providing

Selected Features: Elevation,Horizontal_Distance_To_Roadways,Horizontal_Distance_To_Fire_Points,Soil_Type,Wilderness_Area

Inference:

1) Elevation is the most critical predictive feature in the dataset.
2) Horizontal_Distance_To_Roadways holds the second highest overall predictive importance.
3) Horizontal_Distance_To_Fire_Points is the third most influential feature.

5) Wilderness_Area contributes the least predictive weight among the five.
